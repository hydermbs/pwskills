# **Q1:**
What is Flask Framework? What are the advantages of Flask Framework?
#**Ans:**

Flask is a micro web framework for building web applications in Python. It is lightweight, easy to learn, and provides a simple interface to build web applications quickly. Flask is often used for small to medium-sized projects, such as blogs, small websites, and simple web services.

### **Advantages of Flask Framework include:**

**Lightweight:** Flask is a micro web framework, which means it does not require particular tools or libraries. It is a simple and straightforward framework that provides the necessary tools to build web applications.

**Flexibility:** Flask is highly customizable and flexible. It allows developers to choose and use only the components they need for their project, making it easier to build applications without having to deal with unnecessary features.

**Easy to learn:** Flask is easy to learn and use. It has a minimalistic design and provides a simple interface that makes it easier for developers to get started and build web applications quickly.

**Extensible:** Flask has a modular design that makes it easy to extend with third-party libraries or plugins. This means that developers can add functionality to their application easily without having to reinvent the wheel.

**Large community:** Flask has a large community of developers who contribute to its development and provide support. This means that developers can find a wealth of resources and documentation to help them build web applications more efficiently.

**Scalability:** Flask is scalable and can handle large traffic, making it suitable for building large-scale applications.

Overall, Flask is a simple, flexible, and powerful web framework that allows developers to build web applications quickly and easily. Its popularity is growing rapidly due to its ease of use and flexibility.

# **Q3**
What is App routing in Flask? Why do we use app routes?
#**Ans**
In Flask, app routing is the process of binding URL paths to a specific view function. It defines the URL paths of a Flask web application and maps them to the functions that should handle the requests received from those paths.

Flask uses the @app.route decorator to define routes in the application. The decorator takes the URL path as a parameter, and any HTTP method (GET, POST, PUT, DELETE, etc.) that the route should handle. For example:
We use app routes in Flask to define the behavior of the application when specific URLs are accessed. By defining app routes, we can ensure that the application handles the requests correctly and serves the appropriate responses to the users.

In addition, app routes help to organize the code and make it easier to manage. By separating the code that handles different routes, we can create more modular and maintainable applications.

Overall, app routing is a critical part of Flask web development. It allows developers to define the URLs and behavior of their web applications and make them accessible to users.

route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
from flask import Flask
app = Flask(__name__)

@app.route("/welcome")
def welcome():
  return "Welcome to ABC Corporation"

@app.route("/")
def detail():
  Company_Name = "ABC Corporation"
  Location = "India"
  Contact = "999-999-9999"
  return "Company Name: {}  Location: {}  Contact Number: {}".format(Company_Name, Location, Contact)
if __name__=="__Main__":
  app.run()
```

# **Q5**:
What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

#**Ans:**
In Flask, the url_for() function is used for URL building. It generates a URL for the given endpoint with the method provided. It takes the endpoint name as its first argument and any additional parameters for the URL construction as keyword arguments.

Here is a sample code demonstrating the working of the url_for() function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def show_user_profile(username):
    return 'User %s' % username

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return 'Post %d' % post_id

with app.test_request_context():
    print(url_for('index')) # Outputs: '/'
    print(url_for('show_user_profile', username='Alice')) # Outputs: '/user/Alice'
    print(url_for('show_post', post_id=3)) # Outputs: '/post/3'
```
