**Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**


Seaborn is a popular data visualization library in Python that builds on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five common plots that you can create using Seaborn and their typical uses:

Scatter Plot:

Uses: A scatter plot displays the relationship between two numeric variables. It is useful for visualizing the distribution and correlation between two continuous variables.

Line Plot:

Uses: A line plot is used to visualize the relationship between two continuous variables over a continuous interval. It is commonly used to display trends and patterns over time.

Bar Plot:

Uses: A bar plot represents categorical data using rectangular bars, with the height or length of each bar proportional to the values it represents. It is used to compare the values of different categories and show their relationships.

Histogram:

Uses: A histogram displays the distribution of a single numeric variable by dividing it into bins and counting the number of observations that fall into each bin. It is useful for understanding the shape, central tendency, and spread of the data.

Box Plot:

Uses: A box plot, also known as a box-and-whisker plot, is used to display the distribution of a continuous variable through its quartiles. It shows the minimum, first quartile, median, third quartile, and maximum values, along with any potential outliers. It helps in understanding the spread and skewness of the data.

These are just a few examples of the types of plots that can be created using Seaborn. The library offers many other plot types, such as violin plots, heatmaps, regression plots, and more, providing a wide range of options for visualizing different types of data.

**Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.**

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot(x = fmri.timepoint,y=fmri.signal, hue = fmri.event,style=fmri.region)
```

**Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.**

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.boxplot(x=titanic.pclass, y= titanic.age)
```

```python
sns.boxplot(x='pclass',y=f'fare',data=titanic)
```

**Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.**

```python
diomand = sns.load_dataset('diamonds')
diomand
```

**Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.**

**Que 6: Use the "flights" dataset from seaborn to plot a heatmap.**

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights_pivot = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_pivot)
```
